```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from math import e

def linear(x, *p):
    '''
    Define a fitting function (straight line)
    '''
    return p[0]*x + p[1]

df = pd.read_csv('1.csv')

y = df['Power']
x = df['I']

x_fit = x[285:]
y_fit = y[285:]
# initial guess
p_guess = np.array([0.0, 1.0]) 
# And do the fit:
popt, pcov = curve_fit(linear, x_fit, y_fit, p_guess)  


plt.plot(x_fit, linear(x_fit, *popt), 'r--', label='fitted, slope = 1.14')
plt.plot(x,y, label='P-I curve')
plt.xlabel('Current (mA)')
plt.ylabel('Power (mW)')
plt.title('first 500 um sample, I_th = 588.895')
plt.grid()
plt.legend()

print(linear(588.895, *popt))
print('slope: ' + str(popt[0]))
print('intersect: ' + str(popt[1]))
```
